```python
import nbformat
from nbconvert import PythonExporter
from colorama import Fore, Style

# File descriptions
file_info = {
    'Simulate_Match.ipynb': 'Simulate a Football match between teams',
    'Compare_Players.ipynb': 'Compare attributes of two players',
    'Attributes_&_Ratings.ipynb': 'Explore attributes and compare actual and predicted ratings',
    'Map_and_Stats.ipynb': 'Display geographical data and statistics'
}

# Function to execute a notebook file
def execute_notebook(file_path):
    with open(file_path) as f:
        nb = nbformat.read(f, as_version=4)
        py_exporter = PythonExporter()
        py_code, _ = py_exporter.from_notebook_node(nb)
        exec(py_code, globals())

# Function to display options and handle user input
def display_options():
    while True:
        print(welcome_msg)
        print(f"{Fore.YELLOW}Explore various football-related analyses, match predictions, player comparisons, and more! \n")
        print(f"{Fore.GREEN}Choose an option:")
        for i, file in enumerate(file_info, start=1):
            print(f"{Fore.CYAN}{i}. {file_info[file]}")
        print(f"{Fore.RED}0. Exit{Style.RESET_ALL}")

        try:
            user_input = int(input(f"{Fore.YELLOW}Enter the number of the option you want to run (0 to exit): "))
            if user_input == 0:
                break
            elif 1 <= user_input <= len(file_info):
                selected_file = list(file_info.keys())[user_input - 1]
                execute_notebook(selected_file)
                print(f"{Fore.GREEN}Executed: {file_info[selected_file]}")
                input("Press Enter to continue...")
            else:
                print(f"{Fore.RED}Invalid input. Please enter a number within the provided range.{Style.RESET_ALL}")
        except ValueError:
            print(f"{Fore.RED}Invalid input. Please enter a number.{Style.RESET_ALL}")

# Display a bigger welcome message using ASCII art
welcome_msg = f"""
{Fore.GREEN}
~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~
~██╗    ██╗███████╗██╗      ██████╗ ██████╗ ███╗   ███╗███████╗██╗~
~██║    ██║██╔════╝██║     ██╔════╝██╔═══██╗████╗ ████║██╔════╝██║~
~██║ █╗ ██║█████╗  ██║     ██║     ██║   ██║██╔████╔██║█████╗  ██║~
~██║███╗██║██╔══╝  ██║     ██║     ██║   ██║██║╚██╔╝██║██╔══╝  ╚═╝~
~╚███╔███╔╝███████╗███████╗╚██████╗╚██████╔╝██║ ╚═╝ ██║███████╗██╗~
~ ╚══╝╚══╝ ╚══════╝╚══════╝ ╚═════╝ ╚═════╝ ╚═╝     ╚═╝╚══════╝╚═╝~
~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~

"""

# Start the program by displaying options
display_options()
```
